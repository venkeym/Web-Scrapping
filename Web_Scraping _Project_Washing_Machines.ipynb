```python
url = 'https://www.flipkart.com/search?q=washing+machines&sid=j9e%2Cabm%2C8qx&as=on&as-show=on&otracker=AS_QueryStore_OrganicAutoSuggest_1_11_na_na_na&otracker1=AS_QueryStore_OrganicAutoSuggest_1_11_na_na_na&as-pos=1&as-type=RECENT&suggestionId=washing+machines%7CWashing+Machines&requestId=ba7c024e-2d8c-4a84-8499-d52a21871201&as-backfill=on&page='
```

```python
urls = []
for i in range(1,30):
    urs =url+str(i+1)
    urls.append(urs)
```

```python
urls
```

```python
brand=[]
prices =[]
Discount=[]
Ratings = []
Capacity = []
colour = []
User_Ratings = []
original_price = []
```

```python

for i in urls:
    page = requests.get(i)
    soup= BeautifulSoup(page.text)

    for i in soup.find_all("div",class_="_4rR01T"):
        brand.append(i.text.split()[0])

    for i in soup.find_all("div",class_="_30jeq3 _1_WHN1"):
        a =i.text.split('₹')[-1].replace(',','')
        prices.append(int(a))
    
    for i in soup.find_all("div",class_="_3Ay6Sb"):
        a =i.text.split('%')[0].replace(',','')
        Discount.append(int(a))
    
    for i in soup.find_all("div",class_="_3LWZlK"):
        Ratings.append(float(i.text.split()[0]))
    
    for i in soup.find_all("div",class_="_4rR01T"):
        Capacity.append((i.text.split('kg')[0].split()[-1]))
    
    for i in soup.find_all("div",class_="_4rR01T"):
        colour.append(i.text.split()[-1])
       # colour.append(i.text.split('kg')[0].split()[-1])
    
    for i in soup.find_all("span",class_="_2_R_DZ"):
        User_Ratings.append(int(i.text.split()[0].replace(",","")))
   
    
    for i in soup.find_all("div",class_="_3I9_wc _27UcVY"):
        a =i.text.split('₹')[-1].replace(',','')
        original_price.append(int(a))

    
```
